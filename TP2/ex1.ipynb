```python
import numpy as np
import matplotlib.pyplot as plt

nb_boule = 8
a_tirer = 2
valeurs = list(range(1,nb_boule+1))

def tirage(nb_boule,a_tirer):
    tab = []
    i=0
    while i < a_tirer:
        val = np.random.randint(1,nb_boule+1)
        if val not in tab:
            i+=1
            tab.append(val)
    return tab



nb_experience = 15000
result =[]
theo = []
boules_tirées = []
boules = 0

for i in range(nb_boule):
    result.append(0)
    theo.append((2*(nb_boule-(i+1)))/(nb_boule*(nb_boule-1)))


for i in range(nb_experience):
    boules = np.min(tirage(nb_boule,a_tirer))
    result[boules-1]+=1
    boules_tirées.append(boules)
for el in result:
    result[result.index(el)] = result[result.index(el)]/nb_experience

esp_theo = (nb_boule+1)/3
ecart_theo = round(np.sqrt((nb_boule**2 - nb_boule - 2)/18),2)

print(result)
print(theo)

hX, binsX, _ = plt.hist(valeurs,weights=result,bins=np.arange(1,nb_boule+2)-0.5,rwidth=0.8,label="empirique")
plt.bar(valeurs,theo,width=0.4,color='c',label="théorique")
plt.ylabel("P(X=k)")
plt.xlabel("k")
plt.title(str(nb_boule) + " boules, " + str(nb_experience) + " expériences")
plt.legend(loc="upper right")
plt.text(6.5, 0.21, s = 'E(X) emp. = '+str(round(np.mean(boules_tirées),2)))    #comment fonctionne mean pour pondérer ?
plt.text(6.5, 0.195, s = 'E(X) théo. = '+str(esp_theo))
plt.text(6.5, 0.17, s = ' σ(X) emp. = '+str(round(np.std(boules_tirées),2)))  #comment fonctionne std pour pondérer ?
plt.text(6.5, 0.155, s = ' σ(X) théo. = '+str(ecart_theo))








```

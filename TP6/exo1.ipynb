```python
import numpy as np
import matplotlib.pyplot as plt
import math

#u = x+2y  avec x et y deux exp et para 1 et 2

N = 50000
nu = 1
mu = 2

x = np.random.rand(N)
y = np.random.rand(N)
X = np.zeros((N,1),dtype=float)
Y = np.zeros((N,1),dtype=float)


X = np.log(1-x)/(-nu)
Y = np.log(1-y)/(-mu)

#Partie théorique
x_théo = np.linspace(0,14,N) # a faire marcher
y_théo = np.zeros((N,1), dtype=float)
y_théo = x_théo*np.exp(-x_théo)
esp_théo = 1/nu + 2*(1/mu)
var_théo = 1/(nu**2) + 4*(1/(mu**2))


plt.figure() #affichage de l'histogramme
hX, binsX, _ = plt.hist(X+2*Y,bins=100,rwidth=0.8,label="Empirique",color="g",density=True)
plt.plot(x_théo,y_théo,label="Théorique",color='b')
plt.ylabel("f(u)")
plt.xlabel("u")
plt.legend(loc="upper right")
plt.text(10, 0.3, s = 'Expériences : '+str(N))
plt.text(10, 0.25, s = 'E(U) emp. = '+str(round(np.mean(X+2*Y),2)))
plt.text(10, 0.20, s = 'E(U) théo. = '+str(esp_théo))
plt.text(10, 0.15, s = ' V(U) emp. = '+str(round(np.std(X+2*Y)**2,2)))  
plt.text(10, 0.10, s = ' V(U) théo. = '+str(var_théo))
plt.show()




```

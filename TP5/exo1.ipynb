```python
#code perdu à refaire

import numpy as np
import matplotlib.pyplot as plt
import math


n = 10000

#PARTIE 1

u1 = np.random.rand(n)
u2 = np.random.rand(n)

X = np.zeros((n,1))
Y = np.zeros((n,1))

for i in range(n):
    X[i] = math.sqrt(-2*math.log(u1[i]))*math.cos(2*math.pi*u2[i])
    Y[i] = math.sqrt(-2*math.log(u1[i]))*math.sin(2*math.pi*u2[i])

#Partie théorique
x_théo = np.linspace(-4,4,n) # on génère nos x
y_théo = np.zeros((n), dtype=float)
for i in range(n):
    y_théo[i] = math.exp((-(x_théo[i])**2)/2)/math.sqrt(2*math.pi)

plt.figure() #affichage de l'histogramme
hX, binsX, _ = plt.hist(Y,bins=100,rwidth=0.8,label="Empirique",color="g",density=True)
plt.plot(x_théo,y_théo,label="Théorique",color='b')

plt.figure() #affichage de l'histogramme
hX, binsX, _ = plt.hist(X,bins=100,rwidth=0.8,label="Empirique",color="g",density=True)
plt.plot(x_théo,y_théo,label="Théorique",color='b')



#PARTIE 2
µ = 4
sigma = 3

#Partie théorique
x_théo = np.linspace(-20,20,n) #faudrait faire un truc responsive
y_théo = np.zeros((n), dtype=float)
for i in range(n):
    y_théo[i] = math.exp(-((x_théo[i]-µ)**2)/(2*(sigma**2)))/(sigma*math.sqrt(2*math.pi))


plt.figure() #affichage de l'histogramme
hX, binsX, _ = plt.hist(sigma*X+µ,bins=100,rwidth=0.8,label="Empirique",color="g",density=True)
plt.plot(x_théo,y_théo,label="Théorique",color='b')
plt.ylim(0,0.4)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math

lmbd = 0.3
micro = 0.5
p = 0.65 #  dfrinir

def  gomtrique(p): #on dinit notre loi gomtrique
    cnt = 0
    n = np.random.rand()
    while not(n<p):
        n = np.random.rand()
        cnt += 1
    return cnt



nb_exp = 10000

result_micro = []
result_lmbd = []
result_p = []
val_tho= []
proba_tho_micro = []
proba_tho_lmbd = []
proba_tho_p = []

for i in range(nb_exp): #on gnre nos rsultats exprimentaux
    result_micro.append(gomtrique(micro))
    result_lmbd.append(gomtrique(lmbd))
    result_p.append(gomtrique(p))

for i in range(15): #on gnre nos valeurs thoriques
    val_tho.append(i)
    proba_tho_lmbd.append(((1-lmbd)**(i))*lmbd)
    proba_tho_micro.append(((1-micro)**(i))*micro)
    proba_tho_p.append(((1-p)**(i))*p)

plt.figure() #affichage de l'histogramme de la premire loi go
hX, binsX, _ = plt.hist(result_lmbd,bins=np.arange(0,10+2)-0.5,rwidth=0.8,label="Empirique",density=True)
plt.bar(val_tho,proba_tho_lmbd,width=0.4,label="Thorique",color='g')
plt.xlabel("k")
plt.legend(loc="upper right")
plt.title("Loi de X : Geo(" + str(lmbd)+ ")")
plt.show()

plt.figure() #affichage de l'histogramme de la deuxime loi go
hX, binsX, _ = plt.hist(result_micro,bins=np.arange(0,10+2)-0.5,rwidth=0.8,label="Empirique",density=True)
plt.bar(val_tho,proba_tho_micro,width=0.4,label="Thorique",color='g')
plt.xlabel("k")
plt.legend(loc="upper right")
plt.title("Loi de X : Geo(" + str(micro)+ ")")
plt.show()

plt.figure() #affichage de l'histogramme de la troisime loi go
hX, binsX, _ = plt.hist(result_p,bins=np.arange(0,10+2)-0.5,rwidth=0.8,label="Empirique",density=True)
plt.bar(val_tho,proba_tho_p,width=0.4,label="Thorique",color='g')
plt.xlabel("k")
plt.legend(loc="upper right")
plt.title("Loi de Z = min(X,Y) : Geo("+str(p)+")")
plt.show()


```

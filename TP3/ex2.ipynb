```python
import numpy as np
import matplotlib.pyplot as plt
import math

lmbd = 0.3
micro = 0.5
p = 0.65 # à défrinir

def  géométrique(p): #on déinit notre loi géométrique
    cnt = 0
    n = np.random.rand()
    while not(n<p):
        n = np.random.rand()
        cnt += 1
    return cnt



nb_exp = 10000

result_micro = []
result_lmbd = []
result_p = []
val_théo= []
proba_théo_micro = []
proba_théo_lmbd = []
proba_théo_p = []

for i in range(nb_exp): #on génère nos résultats expérimentaux
    result_micro.append(géométrique(micro))
    result_lmbd.append(géométrique(lmbd))
    result_p.append(géométrique(p))

for i in range(15): #on génère nos valeurs théoriques
    val_théo.append(i)
    proba_théo_lmbd.append(((1-lmbd)**(i))*lmbd)
    proba_théo_micro.append(((1-micro)**(i))*micro)
    proba_théo_p.append(((1-p)**(i))*p)

plt.figure() #affichage de l'histogramme de la première loi géo
hX, binsX, _ = plt.hist(result_lmbd,bins=np.arange(0,10+2)-0.5,rwidth=0.8,label="Empirique",density=True)
plt.bar(val_théo,proba_théo_lmbd,width=0.4,label="Théorique",color='g')
plt.xlabel("k")
plt.legend(loc="upper right")
plt.title("Loi de X : Geo(" + str(lmbd)+ ")")
plt.show()

plt.figure() #affichage de l'histogramme de la deuxième loi géo
hX, binsX, _ = plt.hist(result_micro,bins=np.arange(0,10+2)-0.5,rwidth=0.8,label="Empirique",density=True)
plt.bar(val_théo,proba_théo_micro,width=0.4,label="Théorique",color='g')
plt.xlabel("k")
plt.legend(loc="upper right")
plt.title("Loi de X : Geo(" + str(micro)+ ")")
plt.show()

plt.figure() #affichage de l'histogramme de la troisième loi géo
hX, binsX, _ = plt.hist(result_p,bins=np.arange(0,10+2)-0.5,rwidth=0.8,label="Empirique",density=True)
plt.bar(val_théo,proba_théo_p,width=0.4,label="Théorique",color='g')
plt.xlabel("k")
plt.legend(loc="upper right")
plt.title("Loi de Z = min(X,Y) : Geo("+str(p)+")")
plt.show()


```

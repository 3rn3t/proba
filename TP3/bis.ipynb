```python
import numpy as np
import matplotlib.pyplot as plt
import math

im = plt.imread("ikonos-high-resolution-satellite-image-bora-bora.tif") #on ouvre l'image

imModif = np.zeros(shape=(im.shape[0],im.shape[1],im.shape[2]), dtype=int)

def bombardements(img,bombes,imModif):
    result = 0
    for i in range(bombes):
        x = np.random.randint(im.shape[0])
        y = np.random.randint(im.shape[1])
        if (img[x][y][2]>=52) and (img[x][y][1]<= 1.13*img[x][y][2]) and (img[x][y][0] <= 0.8*img[x][y][2]):
            result+=1
            imModif[x][y][0] = 255
        else:
            imModif[x][y] = [255,255,255]
    return result/bombes,imModif

n_impacts = 100000

pourcentage_corail,iModif = bombardements(im,n_impacts,imModif)

hauteur_relle = 13900

largeur_relle = (im.shape[1]*hauteur_relle)/im.shape[0]
surface_relle_corail = round(largeur_relle*hauteur_relle*pourcentage_corail/10000,2)
print("Il y a " + str(surface_relle_corail) + " hectares de corail")

plt.imshow(iModif)


```

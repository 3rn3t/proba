```python
import numpy as np
import matplotlib.pyplot as plt

def gen_pts(nb):
    return np.random.rand(2,nb)

N = 500
pts = gen_pts(N)


cnt = 0
for i in range(N):
    if (pts[0][i]**2 + pts[1][i]**2) <= 1:
        plt.plot(pts[0][i],pts[1][i],'g+')
        cnt+=1
    else :
        plt.plot(pts[0][i],pts[1][i],'r+')
plt.axis('equal')
print(cnt*4/N)

cnt1 = 0
for i in range(N):
    if (np.log(pts[0][i]) / (pts[0][i]**2)) >= pts[1][i]:
        plt.plot(pts[0][i],pts[1][i],'g+')
        cnt1+=1
    else :
        plt.plot(pts[0][i],pts[1][i],'r+')
plt.axis('equal')
print(cnt1/N) 






    









```

```python
import numpy as np
import matplotlib.pyplot as plt

def lancer_de_6_faces(N):
    return np.random.randint(1,7,N)


tab = lancer_de_6_faces(6)
print(tab)
hX, binsX, _ = plt.hist(tab,bins=np.arange(1, 8) - 0.5,rwidth=0.8)

nb_exp = 10000

def test_jeu(nb_exp):
    cnt_A=0
    cnt_B=0
    for i in range(nb_exp):
        var = lancer_de_6_faces(4)
        if 6 in var:
            cnt_A+=1
        var = lancer_de_6_faces(24)
        var1 = lancer_de_6_faces(24)
        if 12 in var+var1:
            cnt_B+=1
        i+=1   
    return cnt_A,cnt_B
    
print(test_jeu(nb_exp))


```

```python
import numpy as np
import matplotlib.pyplot as plt
import math


a = -2
b = 3
N = 10000
nb_val = 30
val_exp = []

for i in range(N): #on génère les valeurs de notre loi uniforme expérimentale
    val_exp = ((b-a)*np.random.rand(N)+a)



plt.figure() #affichage de l'histogramme
hX, binsX, _ = plt.hist(val_exp,bins=nb_val,rwidth=0.8,label="Empirique",color="g",density=True)
plt.hlines(1/(b-a),a,b,color="b",label="Théorique")
plt.ylabel("Densité de probabilité")
plt.xlabel("x (valeurs de X)")
plt.ylim(0,1)
plt.legend(loc="upper right")
plt.title("Simulation de la loi uniforme sur l'intervalle [" + str(a) + "," + str(b) + "]")
plt.text(0.6*b, 0.8, s = 'E(X) emp. = '+str(round(np.mean(val_exp),2)))
plt.text(0.6*b, 0.7, s = 'E(X) théo. = '+str(round((a+b)/2,2)))
plt.text(0.6*b, 0.6, s = ' σ(X) emp. = '+str(round(np.std(val_exp),2)))  
plt.text(0.6*b, 0.5, s = ' σ(X) théo. = '+str(round(math.sqrt(((b-a)**2)/12),2)))
plt.show()



```

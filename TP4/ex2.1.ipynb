```python
import numpy as np
import matplotlib.pyplot as plt
import math


a = -3
b = 2
n = 5000


#Partie Empirique
y_empirique = np.random.rand(n) #on génère y

for i in range(n):
    y_empirique[i] = math.exp(y_empirique[i]*math.log(6))-4 #on utilise la bijection réciproque de X et on retombe dans [-3,2]


#Partie théorique
x_théo = np.linspace(a,b,n) # on génère nos x
y_théo = np.zeros((n), dtype=float)
esp_théo = round((5-4*math.log(6))/math.log(6),2) #espérance théorique
ecart_théo = round(math.sqrt((5/math.log(6))*(3.5-(5/math.log(6)))),2) #ecart-type théorique
for i in range(n):
    y_théo[i] = 1/(math.log(6)*(x_théo[i]+4)) # on calcule les y


plt.figure() #affichage de l'histogramme
hX, binsX, _ = plt.hist(y_empirique,bins=50,rwidth=0.8,label="Empirique",color="g",density=True)
plt.plot(x_théo,y_théo,label="Théorique",color='b')
plt.ylabel("f(x)")
plt.xlabel("x (valeurs de X)")
plt.legend(loc="upper right")
plt.title("Simulation de la loi (ln(6)(x+4))^-1")
plt.text(-1, 0.5, s = 'Expériences : '+str(n))
plt.text(-1, 0.45, s = 'E(X) emp. = '+str(round(np.mean(y_empirique),2)))
plt.text(-1, 0.4, s = 'E(X) théo. = '+str(round(esp_théo,2)))
plt.text(-1, 0.35, s = ' σ(X) emp. = '+str(round(np.std(y_empirique),2)))  
plt.text(-1, 0.3, s = ' σ(X) théo. = '+str(ecart_théo))
plt.show()



```
